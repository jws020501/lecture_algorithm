```python
#파이썬 리스트 연습  ㅣ 항목접근
test_list = [1,2,3,"one","two","three"]
print(test_list[0])
print(test_list[3])
print(test_list[-1])
print(test_list[-3])

for item in test_list:
    print(item)
```

```python
#파이썬 리스트 연습  ㅣ 삽입과 삭제
test_list = [1,2,3,"one","two","three"]
test_list.insert(3,4)
test_list.append("four")
test_list.remove("one")
item = test_list.pop(0)
print(item)
item = test_list.pop()
print(item)
```

```python
#파이썬 리스트 연습  ㅣ 반복구조
data_list = [23,45,21,14,2,67]
target = 14
index = 0 
while index <= 5:
    if(data_list[index] ==target): break
    index += 1

print(index)

print(data_list.index(target))
```

```python
#파이썬 리스트 연습  ㅣ 리스트에 저장된 숫자 중 가장 큰 숫자
data_list = [23,45,21,14,2,67]
print(max(data_list))
```

```python
#파이썬 리스트 연습  ㅣ 리스트에 저장된 숫자 중 가장 작은 숫자
data_list = [23,45,21,14,2,67]
print(min(data_list))
```

```python
#파이썬 리스트 연습  ㅣ 리스트에 저장된 숫자 중 입력된 숫자보다 큰 갯수
data_list = [23,45,21,14,2,67]
target = int(input())
print(len(list(filter(lambda x: x>target,data_list))))
```

```python
#파이썬 리스트 연습  ㅣ 리스트 전부더하기
from functools import reduce
data_list = [23,45,21,14,2,67]
print(reduce(lambda x,y: x+y,data_list))
```

```python
data_list = [23,45,21,14,2,67]
print(list(reversed(data_list)))
```

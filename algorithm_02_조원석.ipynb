```python
stack = []
#매뉴 출력
while(1):
    print('''    ==================================
      1.Push 2.Pop 3.Peek 4.exit
    ==================================
    ''')
    #입력 받기
    menu = int(input("    연산메뉴 선택 : "))
    #입력에따라 연산 수행
    if(menu==0):
    #0 누를때까지 반복
        print("    종료")
        break
    elif(menu==1):
        acc = int(input("     [Push]추가 할 데이터 입력 :"))
        stack.append(acc)
    elif(menu==2):
         print(f"    [pop] {stack.pop()}")
    elif(menu==3):
         print(f"    [Peek] {stack[len(stack)-1]}")
        
    #스택의 상태 출력
    print(f"    [stack]{stack}")
```

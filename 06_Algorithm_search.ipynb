```python
# 이진탐색
def binary_search_recursive(A,key,low,high):
    if(low>high):
        return -1 #재귀를 계속 돌다가 key 값이 없으면 low가 high를 넘어갈 수 있다 따라서 무한 재귀를 막고 없는 값이 없는 경우를 return
    middle = (low+high)//2 #middle은 2진탐색을 진행할 때에 범위를 구분하기위한 중간 인덱스
    if(key == A[middle]): 
        return middle    ## 찾으려는 수와 데이터의 해당 인덱스의 값과 동일할때 index return
    elif(key<A[middle]):
        return binary_search_recursive(A,key,low,middle-1) ## 찾으려는 key 값보다 데이터의 중간값이 작을 경우 high값을 중간값보다 아래로만 다시 탐색한다
    else:
        return binary_search_recursive(A,key,middle+1,high)## 위와 반대로 key 값보다 데이터의 중간값이 클 경우 low값을 중간값보다 위로만 다시 탐색한다
    
def binary_search_loop(A,key): ## 위와 동일한 프로세스를 진행하나 재귀를 loop로 바꾼 코드입니다 
    low = 0
    high = len(A)-1
    while low <= high:
        middle = (low+high)//2
        if(key == A[middle]):
            return middle
        elif(key<A[middle]):
            high = middle - 1
        else:
            low = middle + 1

data = [2,6,11,13,18,20,22,27,29,30,34,38,41,42,45,47]
num = int(input("찾을 숫자 : "))
index = binary_search_recursive(data,num,0,len(data))
if index == -1:
    print("없는 숫자입니다")
else:    
    print(f"숫자 {num}은 index:{index}에 있습니다")
index = binary_search_loop(data,num)
if index == -1:
    print("없는 숫자입니다")
else:    
    print(f"숫자 {num}은 index:{index}에 있습니다")
    
```
